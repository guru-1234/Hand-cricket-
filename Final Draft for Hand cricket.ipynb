```python
import random 
from pywebio.input import *
from pywebio.output import *

# Determining toss winner  
u = " "
user = select("Choose your side",['Heads','Tails'])
toss = random.choice(['Heads', 'Tails'])

# Toss winner choosing the favor 
if (user == toss):
    put_text("You won the Toss")
    u = select("Choose Bat or Bowl", ['Bat', 'Bowl']) #Toss in the favor of user 
    put_text("You chose"+" "+u)
elif (user != toss):
    put_text("You loss the Toss")
    c = random.choice(['Bat', 'Bowl']) #Toss in the favor of user 
    put_text("Computer choose" + " " + c)
    
    if c == "Bat":
        u = "Bowl"
    else:
        u = "Bat"
        
# feeding the total no. of overs 
overs = int(input("Enter the No. of Overs")) 
put_text("Total no. of overs",overs)
balls = int(overs*6)
cont = True
total = 0
count = 1

# Varibales for the score of players 
total_1 = 0
total_2 = 0

first = ""
second = ""

if u == "Bat":
    first = "User"
    second = "Computer"
else:
    first = "Computer"
    second = "User"

# 1st innings
for ball in range(balls):
    put_text("Score: {} Over: {}.{}".format(total_1, int((ball+1) / 6), ((ball+1) % 6)))
    
    user = int(input("Enter number ({}.{})\n".format(int((ball+1) / 6), ((ball+1) % 6))))
    
    put_text("User:", user, "\n")
                                                     
    comp = (random.randint(1, 6))
    put_text("Computer:", comp, "\n")
    
    if (comp == user):
        put_text("1st innings over", "\n")
        break
    else:
        if u == "Bat": 
            total_1 = total_1 + user
            put_text("Your total is now", total_1)
        else:
            total_1 = total_1 + comp
            put_text("Comp total is now", total_1)
    
put_text("2nd innings started. Required Score:", total_1, "\n")
        
# 2nd innings
for ball in range(balls):
    put_text("Score: {} Over: {}.{}".format(total_2, int((ball+1) / 6), ((ball+1) % 6)))
    
    user = int(input("Enter number ({}.{})\n".format(int((ball+1) / 6), ((ball+1) % 6))))
    put_text("User", user, "\n")
    comp = (random.randint(1, 6))
    put_text("Computer", comp, "\n")
    
    if (comp == user):
        put_text("2nd innings over", "\n")
        
        if total_1 == total_2:
            put_text("Match Drawn")
        if total_1 > total_2:
            put_text(first + " won")
        if total_2 > total_1:
            put_text(second + " won")
            break
            
    else:
                
        if u == "Bat": 
            total_2 = total_2 + comp
            put_text("Comp total is now", total_2)
                
            if total_2 > total_1:
                put_text("2nd innings over", "\n")
                put_text(second + " won")
                break 
        else:
            total_2 = total_2 + user
            put_text("Your total is now", total_2)
                
            if total_2 > total_1:
                put_text("2nd innings over", "\n")
                put_text(second + " won")
                break 
     

```
